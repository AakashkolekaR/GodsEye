# Score Card

```python
page = requests.get('https://m.cricbuzz.com/live-cricket-scorecard/20567/ausw-vs-nzw-10th-match-group-b-icc-womens-world-t20-2018')
```

```python
soup = BeautifulSoup(page.text)
#print(soup)
```

```python
rows = soup.find_all('tr')
```

```python
list_row = []
```

```python
import re

list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '',str_cells))
    print(clean2)
    list_rows.append(clean2)
```

```python
df = pd.DataFrame(list_rows)
df
```

```python
a = ["1.mp4","10.mp4","100.mp4","2.mp4","20.mp4","200.mp4"] 
```

```python
for i in range(len(a) - 1):
    for j in range(len(a) - 1):
        if int(a[j].split('.')[0]) > int(a[j + 1].split('.')[0]):
            t = a[j]
            a[j] = a[j + 1]
            a[j + 1] = t
```
